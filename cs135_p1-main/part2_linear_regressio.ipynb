```python
import pandas as pd
import numpy as np
import os
import sklearn
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
data_dir = 'data_reviews'
y_train_df = pd.read_csv(os.path.join(data_dir, 'y_train.csv'))
x_train_embeddings = np.load('data_reviews/x_train_BERT_embeddings.npy')
x_test_embeddings = np.load('data_reviews/x_test_BERT_embeddings.npy')
y_train_array = y_train_df.to_numpy()
y_train_array = y_train_array.reshape((2400,))  #converting input y df to array
```

```python
lr_model = LogisticRegression()
my_param_grid = dict()
my_param_grid['C'] = np.logspace(-6, 6, 12)
```

```python
y_train_array = y_train_df.to_numpy()
y_train_array = y_train_array.reshape((2400,)) 
```

```python
grid_search = sklearn.model_selection.GridSearchCV(
    lr_model,
    my_param_grid,
    scoring='roc_auc',
    cv=5,
    refit=False)
```

```python
grid_search.fit(x_train_embeddings, y_train_array)
```
